```python
from scipy.special import erfcinv
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
y = []
for i in range (0,30,2):
    ebno = 10**(i/10)
    pb = 15/64*math.erfc(np.sqrt(4/85*ebno))-225/2048*(math.erfc(np.sqrt(4/85*ebno))**2)
    y.append(pb)

ax.semilogy(range(0,30,2),y, 'o-')
ax.axis([0,30,0.99e-10,1])
ax.set(xlabel='Eb/No (dB)', ylabel='Bit error rate (BER)',
    title='QAM 256 (Trade Off)')
ax.grid()

plt.show()
```

```python
b = np.sqrt(4/85*ebno)
a = math.erfc(b)
print(f'Nilai dari erfc {b} (x) adalah {a} (y) dan inverse efrc dari y adalah {erfcinv(a)} = nilai (x)')
```

```python
y = []
xeno = []

for i in np.arange(20, 26,0.5):
    ebno = 10**(i/10)
    pb = 15/64*math.erfc(np.sqrt(4/85*ebno))-225/2048*(math.erfc(np.sqrt(4/85*ebno))**2)
    y.append(pb)
    xeno.append(ebno)

fig, ax = plt.subplots()
ax.semilogy(np.arange(20, 26,0.5),y, 'o-')
ax.axis([20,26,0.99e-10,1])
ax.set(xlabel='Eb/No (dB)', ylabel='Bit error rate (BER)',
    title='QAM 256 (Trade Off)')
ax.grid()
plt.show()
```
